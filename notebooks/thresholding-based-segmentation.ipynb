```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
#read image
image = cv2.imread('../assets/a_on_fire.jpg', 0)
type(image)
```

```python
# shape of the original image
image.shape
```

```python
plt.figure(figsize=(8,8))
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
flattened_image = image.reshape((image.shape[0] * image.shape[1],))
```

```python
#shape of the flattend image
flattened_image.shape
```

## Image instensities

```python
plt.figure()
sns.displot(flattened_image, kde=True)
plt.show()
```

## Applying Otsu Thresholding to the image for segmentation

```python
ret, thresh1 = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(thresh1, cmap='binary')
plt.axis('off')
plt.show()
```
